# MechanicsKit: labeled LaTeX with `la` and `ltx`

This notebook shows the modern API for displaying NumPy arrays and SymPy
expressions as labeled LaTeX. The current entry points are:

- `la` — a renderer for a single array, matrix, or expression. Works as
  a function call (`la(A)`) or pipe (`A | la`), with chainable verbs
  like `.evalf(n)`, `.arraystretch(n)`, `.shape()`.
- `ltx` — compose multiple labels and arrays into one display, useful
  for the "Let A=…, B=…, x=…" pattern.

The legacy `display_labeled_latex` function still works (and is shown at
the end for reference), but new code should prefer `la` and `ltx`.

```python
import numpy as np
from mechanicskit import la, ltx
```

## Example 1: Scalar values

A scalar wrapped in `la` renders inline; `ltx` adds a label.

```python
x = np.array(42.5)
la(x)
```

```python
ltx(r"x &= ", x, precision=1)
```

## Example 2: Vectors

`la` displays a column vector by default. Use `ltx` to attach a label,
and `precision` to control how many decimals are shown.

```python
F = np.array([100.0, 0.0, -50.0])
ltx(r"\mathbf F &= ", F, precision=1)
```

```python
u = np.array([0.1234, 0.5678, -0.9012])
ltx(r"\mathbf u &= ", u, precision=4)
```

## Example 3: Matrices

`la` and `ltx` both accept 2-D NumPy arrays. `arraystretch` controls
the row spacing of the rendered `bmatrix`.

```python
theta = np.pi / 4   # 45 degrees
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)],
])
ltx(r"\mathbf R &= ", R, precision=3)
```

```python
K = np.array([
    [ 200, -100,    0],
    [-100,  200, -100],
    [   0, -100,  200],
])
ltx(r"\mathbf K &= ", K, precision=0, arraystretch=2.0)
```

## Example 4: Larger matrices

A 4×4 transformation matrix, with row spacing for readability.

```python
T = np.array([
    [1, 0, 0, 5],
    [0, 1, 0, 3],
    [0, 0, 1, 2],
    [0, 0, 0, 1],
])
ltx(r"\mathbf T &= ", T, precision=0, arraystretch=1.5)
```

## Example 5: Complex numbers

Useful for eigenvalues from a vibration or buckling problem.

```python
lambdas = np.array([3 + 2j, 3 - 2j, -1 + 0j])
ltx(r"\boldsymbol\lambda &= ", lambdas, precision=2)
```

## Example 6: SymPy symbolic matrices

`la` and `ltx` detect SymPy objects automatically. Use them anywhere a
matrix or expression appears — no separate code path.

```python
import sympy as sp

theta = sp.symbols('theta', real=True)

R_sym = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)],
    [sp.sin(theta),  sp.cos(theta)],
])
ltx(r"\mathbf R(\theta) &= ", R_sym)
```

```python
L_T = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta), 0, 0],
    [sp.sin(theta),  sp.cos(theta), 0, 0],
    [0, 0, sp.cos(theta), -sp.sin(theta)],
    [0, 0, sp.sin(theta),  sp.cos(theta)],
])
ltx(r"\mathbf L^{\mathsf T} &= ", L_T, arraystretch=1.8)
```

## Example 7: Pipe syntax and chainable verbs

Every kwarg has a chainable equivalent. Compose them with `|`.

```python
A = np.array([[1.111, 2.222], [3.333, 4.444]])
A | la
```

```python
# Two decimal places, double row spacing, with shape subscript
A | la(evalf=2, arraystretch=2.0, show_shape=True)
```

```python
# Same thing written with chainable verbs
A | la.arraystretch(2.0) | la.shape()
```

```python
# .simplify() and .trigsimp() apply SymPy simplification before rendering
expr = sp.sin(theta)**2 + sp.cos(theta)**2
expr | la.simplify()
```

## Example 8: Multiple arrays in one line with `ltx`

`ltx` accepts alternating LaTeX strings and array/expression arguments.
It is meant for definitions like "Let A=…, B=…, x=…".

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
x = np.array([1, 2, 3])
ltx(r"A = ", A, r",\ B = ", B, r",\ \mathbf x = ", x)
```

## Example 9: `aligned=True` for multi-row equations

When you want a stack of equations sharing an alignment column, pass
`aligned=True` and use `&` plus `\\` to lay out the rows.

```python
ltx(
    r"\bm v &= \dot{\bm r}",
    r"\\ \bm a &= \ddot{\bm r}",
    aligned=True,
)
```

## Example 10: Array dimensions with `show_shape`

The `show_shape` kwarg (and the `.shape()` chainable) appends an
`m × n` subscript — useful in teaching material so students see the
shape of every operand.

```python
u = np.array([0., 0., 2.0, -7.2, 1.6, -7.6, 0., 0.])
ltx(r"u &= ", u, precision=1, show_shape=True)
```

```python
U = np.array([[0, 0], [2, -7.2], [1.6, -7.6], [0, 0]])
ltx(r"U &= ", U, precision=1, show_shape=True)
```
